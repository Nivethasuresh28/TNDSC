```python
import tensorflow as tf
from transformers import T5Tokenizer, TFT5ForConditionalGeneration

# Load the pre-trained T5 model and tokenizer
tokenizer = T5Tokenizer.from_pretrained("t5-small")
model = TFT5ForConditionalGeneration.from_pretrained("t5-small")

def summarize_text(input_text, max_length=100):
    # Tokenize the input text
    input_ids = tokenizer.encode(input_text, return_tensors="tf", max_length=512, truncation=True)

    # Generate the summary
    summary_ids = model.generate(input_ids, max_length=max_length, num_beams=4, early_stopping=True)

    # Decode the summary
    summary_text = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

    return summary_text

# Example usage
input_text = """
Text summarization is the process of distilling the most important information from a source (or sources) to produce
an abridged version for a particular purpose. It's a challenging task in natural language processing, and there are
various techniques to accomplish it. One popular approach is using deep learning models like the Transformer
architecture. In this example, we'll use the T5 model, which is trained specifically for tasks like text
summarization.
"""

summary = summarize_text(input_text)
print("Summary:")
print(summary)
```
